```python
# TO DO LIST
from tkinter import *
from tkinter import messagebox
import pygame

# window set up
window = Tk()
window.title("MY TO DO LIST :)")
window.minsize(840,480)
window.maxsize(840,640)
window['background'] = '#11CDDD'
pygame.mixer.init()

# the task lists:
unfinished_task_list = []
daily_task_list = ['brush teeth', 'make coffee']
finish_task_list = []

#the many different custom commands:

def add_daily_work(daily_task_list):
    for i in range(len(daily_task_list)):
        unfinished_task_list.append(daily_task_list[i])
        lb.insert(END, daily_task_list[i])
        print("DONE")
        print(unfinished_task_list)

def play():
    pygame.mixer.music.load("Congratulations.mp3")
    pygame.mixer.music.play(loops = 0)
    
def addwork():
    task = input_entry.get()
    if task != "":
        unfinished_task_list.append(task)
        lb.insert(END, task)
        input_entry.delete(0, "end")
        messagebox.showinfo("ACTION RESULT", "Successful input")
    else:
        messagebox.showwarning("warning", "Please enter some task.")
    print(unfinished_task_list)
    
def search_task():
    search = input_entry.get()
    if search in unfinished_task_list:
        messagebox.showinfo("WORK STATUS", f"the task: {search} has not been finished :(")
    elif search in finish_task_list:
        messagebox.showinfo("WORK STATUS", f"the task: {search} has been finished :)")
        
def deleteTask():
    for i in lb.curselection():
        if lb.get(i) in unfinished_task_list:
            unfinished_task_list.remove(lb.get(i))
            print("there is something")
        lb.delete(ANCHOR)
        messagebox.showwarning("WARNING", "YOU HAVE REMOVE SOMETHING!")      
        if len(unfinished_task_list) == 0:
            messagebox.showinfo("CONGRATULATION", "You have finish the to do list for today")
            play()
    
    
def selected_item():
    for i in lb.curselection():
        if lb.get(i) in unfinished_task_list:
            unfinished_task_list.remove(lb.get(i))
            print("there is something")
        lb1.insert(END,lb.get(i))
        finish_task_list.append(lb.get(i))
        print(lb.get(i))
        lb.delete(ANCHOR)
        print(unfinished_task_list)
        if len(unfinished_task_list) == 0:
            messagebox.showinfo("CONGRATULATION", "You have finish the to do list for today")
            play()
            lb1.delete(0, END)

#the list box
frame = Frame(window, background = "red",  relief="sunken", borderwidth=2)
frame1 = Frame(window, background = "green",  relief="sunken", borderwidth=2)
unfinish_lbl = Label(frame, text = "UNFINISHED", font=('Times', 24,'bold'), background = "red")
finish_lbl = Label(frame1, text = "FINISHED", font=('Times', 24,'bold'), background = "green")
button_frame = Frame(window,background = "blue", relief="sunken", borderwidth=2)
sb = Scrollbar(frame)
sb1 = Scrollbar(frame1)

# the two list box
lb = Listbox(
    frame,
    width=20,
    height=6,
    font=('Times', 16),
    bd=0,
    fg='#464646',
    highlightthickness=0,
    selectbackground='#a6a6a6',
    activestyle="none",
    yscrollcommand=sb.set
)
lb1 = Listbox(
    frame1,
    width=20,
    height=6,
    font=('Times', 16),
    bd=0,
    fg='#464646',
    highlightthickness=0,
    selectbackground='#a6a6a6',
    activestyle="none",
    yscrollcommand=sb1.set
    
)

#THE Entry:
Intro = Entry(window, width=30, justify=CENTER, font=('Times', 36,'bold'), background ="#11CDDD",borderwidth=5) 
input_entry = Entry(button_frame, width = 30, font=('Times', 16))
add_btn = Button(button_frame, text = "Add", width = 10, command = addwork)
finish_btn = Button(button_frame, text = "Finish", width = 10, command = selected_item)
del_btn = Button(button_frame, text = "Delete", width = 10, command = deleteTask)
search_btn = Button(button_frame, text = "Search", width = 10, command = search_task)


#getting everything in the page
Intro.insert(0, "THE TO DO LIST")
Intro.pack(padx=50)

unfinish_lbl.pack(side = TOP, padx =10)
finish_lbl.pack(side = TOP, padx = 10)

sb.pack(side=RIGHT, fill=Y)
sb.config(command=lb.yview)

lb.pack(padx = 10 , pady = 10)
frame.pack( side = LEFT)

sb1.pack(side=RIGHT, fill=Y)
sb1.config(command=lb1.yview)

lb1.pack(padx = 10, pady = 10)
frame1.pack( side = RIGHT)

button_frame.pack(side = BOTTOM,pady=20)

input_entry.pack()
add_btn.pack(side = RIGHT)
del_btn.pack(side = RIGHT)
finish_btn.pack(side = RIGHT)
search_btn.pack(side = RIGHT)

# adding the daily work
add_daily_work(daily_task_list)

window.mainloop()
```

```python
from tkinter import *
from tkinter import messagebox

def newTask():
    task = my_entry.get()
    if task != "":
        lb.insert(END, task)
        my_entry.delete(0, "end")
    else:
        messagebox.showwarning("warning", "Please enter some task.")

def deleteTask():
    lb.delete(ANCHOR)
    
ws = Tk()
ws.geometry('500x450+500+200')
ws.title('PythonGuides')
ws.config(bg='#223441')
ws.resizable(width=False, height=False)

frame = Frame(ws)
frame.pack(pady=10)

lb = Listbox(
    frame,
    width=25,
    height=8,
    font=('Times', 18),
    bd=0,
    fg='#464646',
    highlightthickness=0,
    selectbackground='#a6a6a6',
    activestyle="none",
    
)
lb.pack(side=LEFT, fill=BOTH)

task_list = [
    'Eat apple',
    'drink water',
    'go gym',
    'write software',
    'write documentation',
    'take a nap',
    'Learn something',
    'paint canvas'
    ]

for item in task_list:
    lb.insert(END, item)

sb = Scrollbar(frame)
sb.pack(side=RIGHT, fill=BOTH)

lb.config(yscrollcommand=sb.set)
sb.config(command=lb.yview)

my_entry = Entry(
    ws,
    font=('times', 24)
    )

my_entry.pack(pady=20)

button_frame = Frame(ws)
button_frame.pack(pady=20)

addTask_btn = Button(
    button_frame,
    text='Add Task',
    font=('times 14'),
    bg='#c5f776',
    padx=20,
    pady=10,
    command=newTask
)
addTask_btn.pack(fill=BOTH, expand=True, side=LEFT)

delTask_btn = Button(
    button_frame,
    text='Delete Task',
    font=('times 14'),
    bg='#ff8b61',
    padx=20,
    pady=10,
    command=deleteTask
)
delTask_btn.pack(fill=BOTH, expand=True, side=LEFT)


ws.mainloop()
```

```python
# Python3 program to get selected
# value(s) from tkinter listbox
 
# Import tkinter
from tkinter import *
 
# Create the root window
root = Tk()
root.geometry('180x200')
 
# Create a listbox
listbox = Listbox(root, width=40, height=10, selectmode=MULTIPLE)
 
# Inserting the listbox items
listbox.insert(1, "Data Structure")
listbox.insert(2, "Algorithm")
listbox.insert(3, "Data Science")
listbox.insert(4, "Machine Learning")
listbox.insert(5, "Blockchain")
 
# Function for printing the
# selected listbox value(s)
def selected_item():

    for i in listbox.curselection():
        print(listbox.get(i))
 
# Create a button widget and
# map the command parameter to
# selected_item function
btn = Button(root, text='Print Selected', command=selected_item)
 
# Placing the button and listbox
btn.pack(side='bottom')
listbox.pack()
 
root.mainloop()

 
```

```python
def LoginPage():
    login_screen=Tk()
    login_screen.title("Login")
    login_screen.geometry("300x250")
    Label(login_screen, text="Please enter login details").pack()
    Label(login_screen, text="").pack()
    Label(login_screen, text="Username").pack()
    username_login_entry = Entry(login_screen, textvariable="username")
    username_login_entry.pack()
    Label(login_screen, text="").pack()
    Label(login_screen, text="Password").pack()
    password__login_entry = Entry(login_screen, textvariable="password", show= '*')
    password__login_entry.pack()
    Label(login_screen, text="").pack()
    Button(login_screen, text="Login", width=10, height=1).pack()
    login_screen.mainloop()
LoginPage()
```
